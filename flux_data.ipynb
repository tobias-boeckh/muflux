```python
# flux in units of # muons / cm^2 / fb^{-1} / GeV
norm_mu_plus_flux_values = np.array(
    [
        # 0.0519556,
        9.93932,
        23.8132,
        22.7879,
        17.4044,
        11.3768,
        6.86223,
        4.35569,
        2.97028,
        2.28052,
        1.94188,
        1.76636,
        1.62189,
        1.38628,
        1.11405,
        0.919517,
        0.770649,
        0.653526,
    ]
)

norm_mu_plus_flux_unc = np.array(
    [
        # 0.00578206,
        1.09337,
        2.61948,
        2.5067,
        1.91451,
        1.25146,
        0.754859,
        0.479139,
        0.326744,
        0.250871,
        0.21362,
        0.194313,
        0.178422,
        0.152498,
        0.122552,
        0.101154,
        0.0847781,
        0.0718946,
    ]
)

norm_mu_minus_flux_values = np.array(
    [
        # 0.0467556,
        8.95008,
        20.152,
        17.9394,
        13.8852,
        10.3512,
        7.77493,
        5.48533,
        6.18942,
        6.26937,
        5.11118,
        4.50856,
        4.63532,
        5.34069,
        5.43496,
        4.36822,
        3.01851,
        1.87969,
    ]
)

norm_mu_minus_flux_unc = np.array(
    [
        # 0.00521002,
        0.984554,
        2.21674,
        1.97337,
        1.5274,
        1.13864,
        0.855256,
        0.6034,
        0.680849,
        0.689645,
        0.562244,
        0.495955,
        0.509898,
        0.587483,
        0.597853,
        0.480511,
        0.332043,
        0.206772,
    ]
)

norm_mu_minus_flux = unp.uarray(norm_mu_minus_flux_values, norm_mu_minus_flux_unc)
norm_mu_plus_flux = unp.uarray(norm_mu_plus_flux_values, norm_mu_plus_flux_unc)
```

```python
# Calculate muon flux in units of # muons / cm^2 / fb^{-1}
momentum_widths = momentum_bins[1:] - momentum_bins[:-1]
mu_minus_flux = norm_mu_minus_flux * momentum_widths[:-1]
mu_plus_flux = norm_mu_plus_flux * momentum_widths[:-1]
```

```python
mu_common_flux_values = np.array([379.481 + 462.447])
mu_common_flux_unc = 0.11 * mu_common_flux_values
mu_common_flux = unp.uarray(mu_common_flux_values, mu_common_flux_unc)

mu_flux = np.hstack([mu_minus_flux, mu_common_flux, mu_plus_flux[::-1]])
```
